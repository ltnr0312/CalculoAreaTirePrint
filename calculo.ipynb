## Imagem 1

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Carregar a nova imagem
image_path = 'Pneu\imagem1.png'
image = cv2.imread(image_path)

# Converter a imagem para escala de cinza
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Aplicar um limiar (threshold) para binarizar a imagem
_, binary_image = cv2.threshold(gray_image, 1, 255, cv2.THRESH_BINARY)

# Calcular a rea da cor grafite (pixels brancos na imagem binarizada)
grafite_area = np.sum(binary_image == 255)

# Calcular a rea total da imagem
total_area = binary_image.size

# Calcular a proporo da rea de cor grafite
proportion_grafite = grafite_area / total_area

# Mostrar a imagem original e a imagem binarizada para verificao
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Imagem Original')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.subplot(1, 2, 2)
plt.title('Imagem Binarizada')
plt.imshow(binary_image, cmap='gray')
plt.show()

# Mostrar a rea total calculada e a proporo
print(f'A rea da cor grafite : {grafite_area} pixels quadrados')
print(f'A proporo da rea de cor grafite : {proportion_grafite:.4f}')

# Dimenses da imagem em milmetros
largura_mm = 600
altura_mm = 445

# Dimenses da imagem em pixels
largura_px = image.shape[1]
altura_px = image.shape[0]

# Resoluo em pixels por milmetro
resolucao_px_mm = (largura_px / largura_mm, altura_px / altura_mm)

# rea de um pixel em milmetros quadrados
area_pixel_mm2 = (1 / resolucao_px_mm[0]) * (1 / resolucao_px_mm[1])

# Calcular a rea da cor grafite em milmetros quadrados
area_grafite_mm2 = grafite_area * area_pixel_mm2

print(f'A rea da cor grafite  aproximadamente: {area_grafite_mm2:.2f} mm')
```
